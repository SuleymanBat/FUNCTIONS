# positional and keyword arguments

```python
def hello(name,age=41):
    print('My name is {} and age is {}'.format(name,age))
```

```python
hello('Sulo')
```

```python
def hello(*args,**kwargs):
    print(args)
    print(kwargs)
```

```python
hello('Sulo','Chuck',age=40, birthdate=1978)
```

```python
lst = ['Sulo','Chuck']
dict_args = {'age':40,'birthdate':1978}
```

```python
hello(*lst,**dict_args)
```

```python
lst=[1,2,3,4,5,6,7]

def evenoddsum(lst):
    even_sum=0
    odd_sum=0
    for i in lst:
        if i%2==0:
            even_sum=even_sum+i
        else:
            odd_sum=odd_sum+i
    return even_sum,odd_sum
```

# MAP FUNCTION IN PYTHON

```python
def even_or_odd(num):
    if num%2==0:
        return 'The number {} is Even'.format(num)
    else:
        return 'The number {} is Odd'.format(num)
```

```python
even_or_odd(55)
```

```python
lst=[2,4,6,8,5,3,6,4,3,5,7,933,2,22]
```

```python
map(even_or_odd,lst)
```

```python
list(map(even_or_odd,lst))
```

```python
st=[4,5,44,55,45]
```

```python
list(map(even_or_odd,st))
```

```python
# Python program to demonstrate working 
# of map. 

# Return double of n 
def addition(n): 
	return n + n 

# We double all numbers using map() 
numbers = (1, 2, 3, 4) 
result = map(addition, numbers) 
print(list(result)) 
```

```python
# Double all numbers using map and lambda 

numbers = (1, 2, 3, 4) 
result = map(lambda x: x + x, numbers) 
print(list(result)) 
```

# FILTER FUNCTION

```python
def even(num):
    if num%2==0:
        return True
```

```python
# function that filters vowels 
def fun(variable): 
	letters = ['a', 'e', 'i', 'o', 'u'] 
	if (variable in letters): 
		return True
	else: 
		return False


# sequence 
sequence = ['g', 'e', 'e', 'j', 'k', 's', 'p', 'r'] 

# using filter function 
filtered = filter(fun, sequence) 

print('The filtered letters are:') 
for s in filtered: 
	print(s) 
```

```python
lst1=[]
def lst_square(lst):
    for i in lst:
        lst1.append(i*i)
    return lst1
```
